```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 9)
```

```python
X=pd.read_csv('edlich-kmeans-A0.csv')
X.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X['V1'], X['V2'], X['V3'])
```

As the graph shows, the data points are distributed in five "clouds", so the best k should be k=5. 

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=5)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(X['V1'], X['V2'], X['V3'], c=labels)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=500)
```
